### Importaci√≥n de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda_1 = pd.read_csv(url)
tienda_2 = pd.read_csv(url2)
tienda_3 = pd.read_csv(url3)
tienda_4 = pd.read_csv(url4)



```

```python
# Lista de DataFrames de las tiendas (seg√∫n nombres ya cargados en tu notebook)
tiendas = [tienda_1, tienda_2, tienda_3, tienda_4]
```

```python
# Normalizar nombres de columnas
for df in [tienda, tienda2, tienda3, tienda4]:
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
        .str.replace("√°", "a")
        .str.replace("√©", "e")
        .str.replace("√≠", "i")
        .str.replace("√≥", "o")
        .str.replace("√∫", "u")
    )
```

```python
# Exploraci√≥n r√°pida de cada tienda
for i, df in enumerate(tiendas, start=1):
    print(f"\nüìå Tienda {i} - Dimensiones: {df.shape}")
    display(df.head(3))  # Muestra las primeras 3 filas
    print("\nInformaci√≥n general:")
    print(df.info())
    print("\nValores nulos:")
    print(df.isnull().sum())
    print("\nDescripci√≥n num√©rica:")
    display(df.describe())
```

#1. An√°lisis de facturaci√≥n



```python
# Normalizar columnas: min√∫sculas y sin espacios
tiendas = [tienda, tienda2, tienda3, tienda4]
for i in range(len(tiendas)):
    tiendas[i].columns = tiendas[i].columns.str.strip().str.lower()

print("\nüí∞ INGRESO TOTAL POR TIENDA")
for i, df in enumerate(tiendas, start=1):
    ingreso_total = df['precio'].sum()
    print(f"Tienda {i} - Ingreso total: ${ingreso_total:,.2f}")
```

# 2. Ventas por categor√≠a

```python
print("\nüì¶ VENTAS POR CATEGOR√çA")
for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    ventas_categoria = df['categoria_del_producto'].value_counts()
    print(f"\nTienda {i} - Ventas por Categor√≠a:")
    print(ventas_categoria)
```

# 3. Calificaci√≥n promedio de la tienda


```python
print("\n‚≠ê CALIFICACI√ìN PROMEDIO DE LA TIENDA")
for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    promedio_calificacion = df['calificacion'].mean()
    print(f"Tienda {i} - Calificaci√≥n promedio: {promedio_calificacion:.2f}")
```

# 4. Productos m√°s y menos vendidos

```python
print("\nüèÜ PRODUCTOS M√ÅS Y MENOS VENDIDOS")
for i, df in enumerate(tiendas, start=1):
    ventas_productos = df['producto'].value_counts()
    producto_mas_vendido = ventas_productos.idxmax()
    cantidad_mas_vendido = ventas_productos.max()
    producto_menos_vendido = ventas_productos.idxmin()
    cantidad_menos_vendido = ventas_productos.min()

    print(f"\nTienda {i}:")
    print(f"  M√°s vendido: {producto_mas_vendido} ({cantidad_mas_vendido} ventas)")
    print(f"  Menos vendido: {producto_menos_vendido} ({cantidad_menos_vendido} ventas)")
```

# 5. Env√≠o promedio por tienda

```python
print("\nüöö COSTO DE ENV√çO PROMEDIO POR TIENDA")
for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    promedio_envio = df['costo_de_envio'].mean()
    print(f"Tienda {i} - Costo de env√≠o promedio: ${promedio_envio:.2f}")

```

###****Informe final****

```python
# --- 7. Informe Final y Ranking Ponderado ---

# 1. M√©tricas clave
ingresos = [df['precio'].sum() for df in [tienda, tienda2, tienda3, tienda4]]
calificaciones = [df['calificacion'].mean() for df in [tienda, tienda2, tienda3, tienda4]]
envios = [df['costo_de_envio'].mean() for df in [tienda, tienda2, tienda3, tienda4]]

# 2. Normalizaci√≥n (para comparar m√©tricas en la misma escala)
import numpy as np

def normalizar(valores, invertido=False):
    arr = np.array(valores)
    if invertido:  # Para costos, donde menor es mejor
        arr = arr.max() - arr
    return (arr - arr.min()) / (arr.max() - arr.min())

norm_ingresos = normalizar(ingresos)
norm_calificaciones = normalizar(calificaciones)
norm_envios = normalizar(envios, invertido=True)

# 3. Score ponderado (40% ingresos, 40% calificaci√≥n, 20% env√≠o)
scores = 0.4 * norm_ingresos + 0.4 * norm_calificaciones + 0.2 * norm_envios
ranking = sorted(enumerate(scores, start=1), key=lambda x: x[1], reverse=True)

# 4. Resultados destacados
mejor_tienda_ingresos = ingresos.index(max(ingresos)) + 1
mejor_tienda_calif = calificaciones.index(max(calificaciones)) + 1
mejor_tienda_envio = envios.index(min(envios)) + 1
mejor_tienda_global = ranking[0][0]

# 5. Informe final
print("üìä INFORME FINAL")
print("="*50)
print(f"üí∞ Ingresos totales por tienda: {ingresos}")
print(f"‚≠ê Calificaciones promedio por tienda: {calificaciones}")
print(f"üöö Costo de env√≠o promedio por tienda: {envios}")
print()
print(f"üèÜ Mayor ingreso: Tienda {mejor_tienda_ingresos}")
print(f"üèÜ Mejor calificaci√≥n: Tienda {mejor_tienda_calif}")
print(f"üèÜ Env√≠o m√°s barato: Tienda {mejor_tienda_envio}")
print(f"üåü Mejor rendimiento global (ponderado): Tienda {mejor_tienda_global}")
print()
print("üìà Ranking global de tiendas:")
for t, score in ranking:
    print(f"  - Tienda {t}: Score {score:.2f}")
print()
print("üîç RECOMENDACIONES:")
print("- Fortalecer el inventario y marketing en la tienda con mayores ingresos para consolidar liderazgo.")
print("- Revisar estrategias de atenci√≥n al cliente en tiendas con calificaciones m√°s bajas.")
print("- Analizar viabilidad de reducir costos log√≠sticos en tiendas con env√≠os m√°s caros.")
print("- Replicar estrategias de la tienda con mejor balance entre ingresos y calificaci√≥n en las dem√°s.")
print("- Considerar el puntaje ponderado como referencia para inversiones estrat√©gicas.")
```
